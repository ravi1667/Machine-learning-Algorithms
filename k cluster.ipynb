```python
#input and output- no in unlabelled data
#scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(df)
```

```python
X=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

```python
X
```

```python
#train and test-no
from sklearn.cluster import KMeans
```

```python
mymodel = KMeans(n_clusters=5)
```

```python
clusters=mymodel.fit_predict(X)
```

```python
clusters
```

```python
#elbow method
#ssd=sum of squared distance
```

```python
ssd=[]
```

```python
for k in range(2,20):
    mymodel=KMeans(n_clusters=k)
    mymodel.fit(X)
    ssd.append(mymodel.inertia_)
```

```python
ssd
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=range(2,20),y=ssd)
```

```python
mymodel=KMeans(n_clusters=16)
```

```python
clusters=mymodel.fit_predict(X)
```
