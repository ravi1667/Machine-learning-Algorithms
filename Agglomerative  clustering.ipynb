```python
import pandas as pd
```

```python
df=pd.read_csv('engine.csv')
```

```python
df
```

```python
df=df.drop(columns=['name'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
X=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

```python
X
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
mymodel=AgglomerativeClustering(n_clusters=None,distance_threshold=0)
```

```python
mymodel.fit(X)
```

```python
cluster=mymodel.fit_predict(X)
```

```python
cluster
```

```python
mymodel.children_
```

```python
from scipy.cluster import hierarchy
```

```python
d=hierarchy.linkage(mymodel.children_)
```

```python
d
```

```python
data=pd.DataFrame(d,columns=['one cluster','Another cluster','Distance','no of data points'])
```

```python
data
```

```python
data['Distance'].mean()
```
